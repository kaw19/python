### 1.1 Chamada de uma Função
No contexto da programação, função é um conjunto de instruções que realiza uma computação. Quando você define uma função, você especifica o nome, parâmetro(s) e o conjunto de instruções. Posteriormente, pode-se "chamar" a função pelo seu nome e o argumento(s). A expressão em parênteses é chamada de argumento(s) da função na sua chamada, e de parâmetro(s) da função na sua declaração (definição). O argumento é um valor ou variável que passamos à função como entrada.

### 1.5 Geração de Números Aleatórios
Os algoritmos disponíveis não são capazes de gerar números aleatórios mas apenas números pseudoaleatórios, e são assim denominados por serem gerados por uma computação determinística.

O módulo `random` oferece funções que geram números pseudoaleatórios, tais como a função `random()` que retorna um número real (ponto flutuante, *`float`*) entre 0.0 e 1.0 (incluindo o limite inferior, 0.0, mas não
o limite superior, 1.0). Todas as vezes que chamarmos essa função, teremos o próximo número de uma longa série de números pseudoaleatórios.  

Exemplo:

````python
# import random
# import random as rnd
from random import random

for i in range(10):
    x = random()           # random.random()      # rnd.random()
    print(x, end=', ')
````

A função `random()` é apenas uma das funções do módulo **random** que geram números aleatórios. Outra 
função usual é a `randint()` que recebe dois parâmetros:  `'menor'` e `'maior'` e retorna um inteiro entre `'menor'` e `'maior'`, incluindo os limites.

````python
# Aposta na Mega-Sena
from random import randint
aposta = []
dez = 0
while dez < 6:
    num = randint(1, 60)
    if num not in aposta:
        aposta.append(num)
        dez += 1
    # print(f'{randint(1, 60)}', end=' ')     # podem aparecer valores repetidos
print(aposta)
````

Outra função do módulo `random` muito útil é a `choice()`, que escolhe aleatoriamente um elemento de uma sequência fornecida como argumento:

````python
from random import choice

t = [10, 20, 30, 40, 50, 60]
for esc in range(1,11):
    print(f'Escolha {esc:2d}: {choice(t)}')
````

O módulo `random` também disponibiliza funções para gerar valores aleatórios de acordo com funções de distribuição contínuas: gaussiana, exponencial, gamma e outras.

### 1.6 Funções definidas pelo Usuário
Até agora, usamos apenas as funções do Python, mas também é possível adicionar novas funções. A definição de uma função pelo usuário começa pela especificação do nome da função, parâmetro(s) (se existir(em)) e o conjunto de comandos que serão executados quando a função for chamada.  

Sintaxe:  
``` python
def nome_funcao([parametro1[, parametro2, ...[, parametroN]]]):
    comando (ou bloco de comandos)
```
Uma vez definida, a função pode ser utilizada repetidas vezes ao longo do programa.  

**Exemplo**:

````python
def refrao():
     print('Fica sempre um pouco de perfume\n' \
           'nas mãos que oferecem rosas\n'     \
           'nas mãos que sabem ser generosas')
````

A palavra-chave **`def`** indica a definição de uma função de usuário. O nome da função é `refrao`. As regras para nomes de funções são as mesmas usadas para nominar variáveis. Não se pode usar uma palavra-chave como nome de função, e deve-se evitar também uma variável e uma função com mesmos nomes.  

Os parênteses vazios após o nome da função indicam que essa função não usará argumentos ao ser chamada.
A primeira linha da definição da função é chamada de **cabeçalho** da função, e o restante é o **corpo** da função. O cabeçalho é finalizado sempre com um caracter de dois pontos, `:`, e o corpo tem que estar
indentado. Por convenção, a indentação é de quatro espaços em branco. 

Todos os elementos da linguagem Python são objetos, inclusives as funções de usuários.

````python
print(refrao, type(refrao))
refrao()
````

### 1.7 Fluxo de Execução
Para garantir que a função seja definida antes de sua primeiro chamada, é preciso saber a ordem em que cada declaração é executada -- fluxo de execução.  

A execução sempre começa na primeira declaração do programa. Declarações são executadas uma por vez, na ordem de cima para baixo. Definições de função não alteram o fluxo da execução do programa, e lembre-se de que comandos dentro da função não são executados até que a função seja chamada. 

Uma chamada de função provoca um desvio no fluxo da execução. Em vez de ir para a próxima declaração, o fluxo de execução é desviado para o corpo da função, e depois de ter executado todos os comandos da função, o fluxo de execução volta para o ponto de onde havia sido desviado.

Algumas das funções internas já usadas necessitavam de argumentos. Por exemplo, a função `math.sin()` precisa de um número como argumento, o ângulo (em radianos). Algumas funções requerem mais de um argumento: `math.pow()` recebe dois argumentos, a base e o expoente.

### 1.9 Funções e Subrotinas
Alguns subprogramas, como as funções matemáticas, retornam resultados ao chamador da `função`; outros subprogramas, como `mostra_duas_vezes()`, executam uma ação mas não retornam um valor ao chamador. Elas são chamadas de `subrotinas` ou `funções vazias`.  

Quando se chama um subprograma do tipo `função`, espera-se um valor de retorno; por exemplo, podemos atribuir o retorno da função a uma variável ou usá-lo como parte de uma expressão:

````python
x = math.cos(theta)
y = (math.sqrt(5) + 1) / 2
print(f'x={x} y={y}')
````

Para retornar um resultado de uma função ao chamador, usa-se o comando **`return`** na função criada pelo usuário. Por exemplo, vamos criar uma função para calcular e retornar a soma de dois números.

````python
def soma(x, y):                  # função de soma
    return x + y

print('33 + 54 =', soma(33, 54))
````

Os parâmetros podem ter valores padrões, caso o usuário não queira especificar argumentos usuais nas chamadas às funções.  
Exemplo:

````python
def familia(nome, sobrenome = "da Silva"):     # subrotina
    print(f'{nome} {sobrenome}')

familia("João")
familia("Tobias","Oliveira")
familia("Maria")
````

### 1.10 Número Variável de Argumentos (Python 3)
Na definição de uma função, os **parâmetros** são entidades nomeadas que especificam os **argumentos** que uma determinada função pode aceitar.

Ao programar, você pode não estar ciente de todos os casos de uso possíveis de seu código e pode querer oferecer mais opções para futuros programadores que usem o seu módulo ou para usuários que interagem com o código. Podemos passar um **número variável de argumentos** ao chamar uma função, usando **\*args** e **\*\*kwargs** em nosso código.

Vejamos uma típica função Python que usa dois argumentos:

````python
def multiplicacao(x, y):       # dois argumentos nominados: x, y
    print (x * y)

multiplicacao(6, 9)
````

E se, mais tarde, precisarmos multiplicar três números em vez de apenas dois? Se tentarmos adicionar um novo número à função, conforme mostrado abaixo, receberemos uma mensagem de erro.

````python
multiplicacao(6, 9, 3)
````

#### 1.10.1 Usando \*args
Portanto, prevendo a possibilidade de se usar mais argumentos posteriormente, podemos usar **\*args** como parâmetro da função `multiplicacao()`, da seguinte forma:

````python
def multiplicacao(*args):        # lista de argumentos não nominados
    resultado = 1
    for argumento in args:
        resultado *= argumento
    print(resultado)

multiplicacao(6, 9, 3)
multiplicacao(6, 9, 3, 2, 0.5)
````

Como usamos **\*args** para enviar uma lista de argumentos de comprimento variável para função `multiplicacao()`, fomos capazes de passar tantos argumentos quantos desejamos nas chamadas à função, ou seja, foi possível a criação de um código mais flexível que aceita uma quantidade variada de argumentos não nominados na função em questão.

#### 1.10.2 Usando \*\*kwargs
A forma que usa asterisco duplo, **\*\*kwargs** é usada para passar um dicionário de argumentos de comprimento variável nominados para uma função. Os dois asteriscos (`**`) são os elementos importantes na sintaxe, visto que a palavra **kwargs** é usada convencionalmente, embora não seja imposta pela linguagem.

Assim como **\*args**, **\*\*kwargs** pode aceitar quantos argumentos você quiser. No entanto, **\*\*kwargs** difere de **\*args** porque você precisa atribuir os valores dos argumentos aos respectivos nomes dos parâmetros.

Vamos usar uma função curta para mostrar os argumentos **\*\*kwargs** passados à função:

````python
def mostra_argumentos(**kwargs):    # dicionário de argumentos nominados
    print(kwargs)
    
mostra_argumentos(arg1="Maria", arg2=242.7, arg3=False)
````

**Obs**.: Dependendo da versão do Python 3 que você está usando, o tipo de dados dicionário pode estar não ordenado. No Python 3.6 e superior, você receberá os pares de **valores-chave** em ordem, mas nas versões anteriores, os pares serão exibidos em uma ordem aleatória.

Vamos criar outra função curta para mostrar como usar **\*\*kwargs**, uma função para saudar pessoas cujos nomes (argumentos) são passados como dicionário de nomes:

````python
def mostra_nomes(**kwargs):
    for chave, valor in kwargs.items():
        print("O valor de '{}' é {}".format(chave, valor))

mostra_nomes(nome1="João", nome2="Maria")
````

Vamos agora passar argumentos adicionais à função para mostrar que **\*\*kwargs** aceita quantos argumentos você quiser incluir:

````python
mostra_nomes(nome1="João", 
             nome2="Maria",
             nome3="José",
             nome4="Davi",
             nome5="Matheus")
````

Usar **\*\*kwargs** nos fornece flexibilidade para usar uma quantidade variável de argumentos nominados em nosso programa, ou seja, não é preciso saber quantos argumentos serão passados à uma função, durante a sua declaração.  

#### 1.10.3 Argumentos Ordenados
Ao indicar os argumentos na declaração ou chamada de uma função, é preciso que eles apareçam em uma ordem específica:

<ol><li>Argumentos posicionais formais.</li>
    <li><strong>*args</strong>.</li>
    <li>Argumentos nominados.</li>
    <li><strong>**kwargs.</strong></li></ol>
    
Vejamos um exemplo de função com parâmetros posicionais explícitos junto com **\*args** e **\*\*kwargs**:
```python
def exemplo1(arg1, arg2, *args, **kwargs):
...
```
Usando parâmetros posicionais com parâmetros nomeados, além de **\*args** e **\*\*kwargs**, teríamos:
```python
def exemplo2(arg1, arg2, *args, kw1="Maria", kw2=33.21, **kwargs):
...  
```  
É importante manter a ordem dos argumentos em mente ao criar funções para evitar erros de sintaxe em seu código Python.

#### 1.10.4 Usando **\*args** e **\*\*kwargs** nas Chamadas de Funções
Vamos ver um exemplo usando  **\*args**:

````python
def exemplo3(arg1, arg2, arg3):
    print("arg1: ", arg1, ", arg2: ", arg2, ", arg3: ", arg3, sep='')

argumentos = ("Maria", "João", 33.21)       # tupla de argumentos a serem usados na função 'exemplo3'
exemplo3(*argumentos)
outros = [3.4, True, 'Olá!']
exemplo3(*outros)
````

Da mesma forma, os argumentos **\*\*kwargs**  podem ser usados para chamar uma função. Vamos configurar uma variável igual a um dicionário com 3 pares de valores-chave (usaremos **kwargs** aqui, mas pode ser chamado como você quiser) e passá-lo para uma função com três argumentos:

````python
kwargs = {"arg1": "Maria", "arg2": "José", "arg3": "Pedro"}
exemplo3(**kwargs)
````

**Concluindo**: Podemos usar a sintaxe especial de **\*args** e **\*\*kwargs** na definição de uma função para passar um número variável de argumentos à função. 
A criação de funções que aceitam **\*args** e **\*\*kwargs** é melhor usada em situações em que você espera que o número de entradas na lista de argumentos permaneça relativamente pequeno.

### 1.11 Aplicação: Catenária

Em geometria, é o nome dado à curva plana que representa a forma de equilíbrio de um fio (corda, corrente, cabo) homogêneo, flexível, pesado, suspenso por suas extremidades a partir de dois pontos fixos, e submetido exclusivamente à força da gravidade, curva essa calculada pela função do cosseno hiperbólico.

Dos muitos objetos matemáticos que foram estudados e descritos, a **catenária** (do latim:  cadeia de elos, corrente) é muito útil aos desenvolvedores de jogos eletrônicos (*videogames*) e aos engenheiros eletricistas ao projetarem redes de transmissão de energia elétrica por fios.

#### 1.11.1 Definição Formal
A catenária tem uma equação bem definida, baseada no cosseno hiperbólico:
$$y(x)=a.\cosh\left(\frac{x}{a}\right) \tag 1$$
onde: 
$a$ - altera a "abertura" da curva. No entanto, todas as curvas de catenária são semelhantes, pois são todas versões escalonadas umas das outras. Veja o traçado para vários valores desse parâmetro:

````python
from numpy import cosh, linspace
from matplotlib.pylab import plot, legend, ylim, grid

def catenaria(x,a):
    return a*cosh(x/a)
    
x = linspace(-5,5,1000)
for a in range(1,5):
    y = catenaria(x,a)
    plot(x,y,label=f'catenária a={a}')
ylim(0,20); legend(); grid(True);
````

Enquanto as funções **sen()** e **cos()** são definidos sobre um círculo, seus parceiros hiperbólicos, **senh()** e **cosh()** são definidos sobre uma hipérbole (veja a animação seguinte)<img src="img/animacao_hiperbolica.gif" width=300 style="float:right; width:10px">.

$$\rm{senh}(x)=\frac{e^x-e^{-x}}{2}$$&nbsp;&nbsp;&nbsp;&nbsp;$$\cosh(x)=\frac{e^x+e^{-x}}{2}$$

As catenárias também possuem outra propriedade interessante. Elas permitem aos quadrados se deslocarem sem fazer com que seu centro oscile para cima ou para baixo (veja a animação seguinte)<img src="img/quad_deslocando.gif" width=300 style="float:right; width:10px">.

### 2.1 Abrindo um Arquivo
Quando queremos ler ou escrever um arquivo, primeiro temos que abrí-lo para que o sistema operacional providencie o acesso aos dados do arquivo armazenado no dispositivo de memória secundária. No exemplo seguinte, vamos abrir o arquivo chamado 'caixa.txt' com a opção de anexar novas informações textuais ao arquivo, ou seja, o modo de abertura será 'at', onde 'a' é para anexação de dados e 't' indica dados do tipo textual.

A função `open()` e demais funções relacionadas fazem parte da <a href="#anexo">biblioteca padrão do Python</a> (vem embutida na distribuição do Python, e é bem ampla, oferecendo uma variedade de recursos). 

Sempre que seu código abrir um arquivo para leitura ou gravação, você deve tomar cuidado para que o arquivo seja fechado ao terminar de usá-lo, para que seus dados não fiquem vulneráveis a modificações. Python fornece um gerenciador de contexto `open`, que é projetado para garantir que um arquivo seja fechado mesmo no caso de seu código gerar um erro.

Ajuda mostrada pelo comando: ` >>> help(open)`
>**`open(file, mode='r', buffering=-1, encoding=None, errors=None, newline=None, closefd=True, opener=None)`**  
   Abre o arquivo e retorna um descritor de fluxo (*stream*). Lança um OSError em caso de falha.  

> **'file'** ou é um texto ou uma string de bytes que fornece o nome (e o caminho se o arquivo não estiver no diretório de trabalho atual) do arquivo a ser aberto ou um descritor de arquivo do tipo inteiro. (Se um descritor de arquivo é fornecido, ele é fechado quando o objeto de E/S retornado é fechado, a menos que **'closefd'** seja definido como *False*).  

> **'mode'** é uma string opcional que especifica o modo em que o arquivo é aberto. O padrão é 'r', o que significa aberto para leitura em modo texto. Outros valores comuns são 'w' para escrita (truncar o arquivo se já existe), 'x' para criar e gravar em um novo arquivo, e 'a' para anexar (que em alguns sistemas Unix, significa que todas as gravações serão anexadas ao final do arquivo, independentemente da posição de busca atual). No modo de texto, se a codificação não for especificada, a codificação usada é dependente da plataforma: locale.getpreferredencoding (False) é chamado para obter a codificação local atual. (Para ler e escrever bytes brutos use o modo binário e não especifique a codificação). Os modos disponíveis são:

Especificar `mode = 'w'` indica que iremos gravar no arquivo -- se o arquivo já tiver algum conteúdo, esse conteúdo será apagado antes de ser gravado. A seguir estão os "modos" disponíveis de abertura de um arquivo:

|Caracter | Significado |
| :-----: | :---------- |
|    'r'  | aberto para leitura (padrão) |
|    'w'  | abre para escrita, truncando o arquivo caso já exista |
|    'x'  | cria um novo arquivo e o abre para escrita |
|    'a'  | aberto para escrita, acrescentado ao final do arquivo, se existir |
|    'b'  | modo binário |
|    't'  | modo de texto (padrão) |
|    '+'  | abre um arquivo em disco para atualização (leitura e gravação) |


O modo padrão é 'rt' (aberto para leitura de texto). Para acesso binário aleatório, o modo 'w+b' abre e trunca o arquivo para 0 bytes, enquanto 'r+b' abre o arquivo sem truncamento. O modo 'x' implica 'w' e gera um `FileExistsError` se o arquivo já existe.

Por padrão, os arquivos do tipo **texto** utilizam a especificação de decodificação/codificação unicode (utf-8). Ou seja, quando você lê dados de arquivos com `mode = 'r'`, o Python decodificará automaticamente os dados binários que foram armazenados de acordo com o código **utf-8**, que converte os dados binários em texto escrito armazenado como uma string. Da mesma forma, gravar uma string num arquivo nos modos `'w', 'a', 'x'` ou `'+'` presumirá que a string será codificada em uma representação binária (que é necessária para ser armazenada num arquivo) de acordo com o esquema de codificação **utf-8**.

Podemos forçar o Python a ler e escrever estritamente em termos de dados binários, adicionando um `'b'` aos modos: `'rb', 'wb', 'ab', 'xb', '+b'`. Se você estiver salvando uma matriz NumPy, vc deve abrir um arquivo nos modos `'wb'` ou `'xb'` para gravar dados binários; obviamente, não estamos salvando texto quando salvamos uma matriz NumPy de números.

### 2.3 Lendo Arquivos de Texto
Depois de aberto o arquivo com a função `open()`, temos uma alça para manipular o conteúdo do arquivo em questão. Para ler o conteúdo do arquivo, podemos fazê-lo por caracter (um tanto quanto demorado) ou por linha (mais ágil). O exemplo seguinte mostra a leitura por linha:

````python
# ajuste o caminho do arquivo baixado, na variável 'caminho'
caminho = 'dat/'
arq = open(caminho+'mbox-short.txt')     # open() retorna uma alça (handler) para manipulação de dados do/pro arquivo
linhas = 0
for linha in arq:                        # a iteração de linhas provoca a leitura dos dados no arquivo
    linhas += 1
print('Qtde de linhas no arquivo de emails:', linhas)
arq.close()
````

O método **`close()`** de um objeto arquivo libera todas as informações não gravadas e fecha o objeto arquivo, após o que nenhuma outra gravação pode ser feita.  

O fechamento dos arquivos pode se dar de forma automática se usarmos `try/finallly`:
1. ``` python
arq = open('entrada.txt')
try:
     # processamento do arquivo 
finally:
     arq.close()
```
2. ``` python
with open('entrada.txt') as arq:
    # processamento do arquivo
```

A instrução `with` cuida automaticamente do fechamento do arquivo quando o bloco `with` é encerrado, e mesmo em casos de erro. Recomenda-se o uso da declaração `with`, tanto quanto possível, pois ela permite um código mais limpo e facilita a manipulação de erros inesperados. A instrução **with** é utilizada para garantir a finalização de recursos adquiridos.

Para mais detalhes sobre manipulação de arquivos, visite https://www.codigofluente.com.br/aula-20-python-biblioteca-padrao-modulo-io/.

````python
# com try e finally
try:
    f = open("teste1.txt", "w")
    f.write("Testando escrita...")
finally:
    f.close()

# com with
with open("teste2.txt", "w") as f:
    f.write("Testando escrita com 'with'...")
````

````python
# demonstração do uso do método `read` do objeto 'file'
with open(caminho_do_arq, mode="r") as arq:
    # lê o conteúdo todo do arquivo numa string
    conteudo = arq.read()
print(conteudo)
````

Se o tamanho do arquivo é relativamente pequeno comparado ao da memória principal, então podemos ler o arquivo todo em uma string usando o método `read()` do identificador de arquivo, como mostra o exemplo seguinte:

````python
arq = open(caminho+'mbox-short.txt')
conteudo = arq.read()                       # armazena todos os caracteres do arquivo na var. 'conteudo'
print('Qtde de caracteres:', len(conteudo))
print(conteudo[:50])
if arq.closed == False:
    arq.close()
````

Quando o arquivo é lido dessa forma, todos os caracteres, incluindo os caracteres de novas linhas (*newline*) formam uma grande string armazenada na variável `conteudo`. É uma boa ideia armazenar a saída de `read()` numa variável porque cada chamada de `read()` faz uso do recurso computacional.

#### Métodos do Objeto `file`:

* **read()**: Lê todo o conteúdo do arquivo como uma string ou como bytes (dependendo do modo de abertura)
* **readline()**: Lê a próxima linha de texto do arquivo, incluindo o caractere final '\n'

Além disso, é importante observar que o objeto `file` pode ser iterado e que cada iteração retornará uma linha individual de texto do arquivo. Esta é a melhor maneira de ler um arquivo inteiro, linha a linha.

### 2.4 Busca em um Arquivo
Quando se está procurando por dados em um arquivo, é muito comum percorrer o arquivo, ignorando a maioria das linhas e apenas processando as que combinam com uma condição particular. Podemos ajuntar a leitura do arquivo com os métodos de strings para construir um mecanismo de busca.  

Por exemplo, se quisermos ler um arquivo e mostrar apenas as linhas que começam com o prefixo "From:", podemos usar o método `startswith()` para selecionar apenas as linhas que possuem o prefixo desejado. Usamos também o método `rstrip()` para eliminar os espaços em branco à esquerda ou direita da string:

````python
arq = open(caminho+'mbox-short.txt')
for linha in arq:
    if linha.startswith('From:'):
        print(linha.rstrip())
````

Vamos usar o método `find()` de strings para procurar por uma ocorrência de uma string dentro de outra e retornar a posição dela ou -1 se não for encontrada. Em particular, vamos mostrar somente as linhas que contém a string “@uct.ac.za” (isto é, emails oriundos da Universidade de Cape Town, na África do Sul):

````python
arq = open(caminho+'mbox-short.txt')
for linha in arq:
    if linha.startswith('From:'):
        if linha.find('@uct.ac.za') == -1: continue
        print(linha.rstrip())
````

No script anterior usamos a forma contraída da sentença `if` quando colocamos o `continue` na mesma linha do `if`. Essa forma funciona do mesmo jeito que se o `continue` estivesse na próxima linha e indentado.

```python
# corresponde a qualquer coisa que comece com `arq` e termine com `.txt` como

O módulo `pathlib` fornece funcionalidade conveniente para agrupamento de arquivos (*globbing*). ````python
# agrupa todos os arquivos de texto do diretório atual que inicia com 'arq' e termina com '.txt'
from pathlib import Path
dir_corrente = Path('.')
arqs = dir_corrente.glob('arq*.txt')  # essa chamada produz um gerador
arqs
````

### 3.1 Introdução 
Queremos evitar situações desagradáveis, como a perda de dados calculados em seu programa Python. Então, vamos mostrar como **salvar dados de maneira fácil** em arquivos sequenciais, de modo que seu programa possa recupera-los posteriormente. Vamos <font color=red><b>'conservar'</b></font> (*pickling*) os dados, para que eles não se percam...

Python oferece um módulo para essa finalidade, chamado `pickle`. Com os métodos do módulo **pickle** podemos serializar/desserializar estruturas de objetos Python. **'Conservar'** (*pickling*) é o processo de converter uma hierarquia de objetos Python num fluxo de bytes: **objetos --> fluxo de bytes**, e **'desconservar'** (*unpickling*) é a operação inversa, isto é, o fluxo de bytes é convertido novamente em uma hierarquia de objetos Python: **fluxo de bytes --> objetos**. O que chamamos de conservação é também conhecido como 'serialização' ou 'achatamento' de uma estrutura de dados; e desconservação, como 'desserialização'.

### 3.2 Métodos
Um objeto pode ser **serializado** (achatado) com o método de despejo (*dump*) do módulo `pickle`:

```python
pickle.dump(obj, arquivo[, protocolo, *, fix_imports=True])
```
`dump()` cria uma representação serializada de `obj` para salvar os dados do objeto num arquivo sequencial. O argumento facultativo `protocolo` diz ao serializador para usar um dado formato:

* **versão 0** é o formato original (antes do Python3) legível (ASCII) e compatível com versões anteriores do Python.
* **versão 1** é o formato binário antigo que também é compatível com versões anteriores do Python.
* **versão 2** foi introduzido na versão Python2.3, oferece conservação mais eficiente de classes de estilos novos.
* **versão 3** foi introduzido na versão Python3.0, tem suporte explícito para bytes e não pode ser desserializado (*unpickled*) pelo módulo *Pickle* do Python2.x. É o formato recomendado para Python3.x.

O protocolo padrão de Python3 é a **versão 3**.

Se `fix_imports` for `True` e o protocolo for menor que 3, *Pickle* tenta mapear os novos nomes do Python3 para os nomes do módulo antigo utilizados no Python2, de modo que o fluxo de dados *Pickle* seja legível com Python2.

Os objetos descarregados (com `pickle.dump`) num arquivo podem ser relidos num programa usando o método `pickle.load(arquivo)`. Esse método reconhece automaticamente o formato usado na escrita dos dados. 

#### Exemplo:

````python
import pickle
cidades = ['Pirenópolis', 'Corumbá', 'Luziânia', 'Cidade de Goiás', 'Cavalcante']
fh = open('dados.pkl', 'wb')
pickle.dump(cidades, fh)
fh.close()
````

O arquivo de "**dados.pkl**" pode ser lido novamente por Python, na mesma, ou em outra sessão, ou por um programa diferente:

````python
f = open("dados.pkl", "rb")
cities = pickle.load(f)
print(cities)
````

Apenas os objetos, não os seus nomes, são guardados. É por isso que usamos a atribuição de `cities` no exemplo anterior, ou seja, `cities = pickle.load(f)`.

O exemplo anterior conservou apenas um objeto: uma lista das cidades goianas históricas. Mas como seria a consevação de vários objetos? Embale os objetos em um único objeto. Vamos embalar duas listas, **'linguagens_de_programação'** e **'dialetos_python'**, em uma única tupla chamada **'objeto_pickle'**, no exemplo a seguir:

````python
arq = open("dados2.pkl","bw")
linguagens_de_programação = ["Python", "Perl", "C++", "Java", "Lisp", "JavaScript"]
dialetos_python = ["Jython", "IronPython", "CPython"]
objeto_pickle = (linguagens_de_programação, dialetos_python)
pickle.dump(objeto_pickle, arq)
arq.close()
````

A conserva de dados do exemplo anterior, isto é, os dados gravados no arquivo `dados2.pkl`, pode ser separada em duas listas, novamente, quando relermos os dados:

````python
arq = open('dados2.pkl','rb')
(lings, dialetos) = pickle.load(arq) 
print(lings, dialetos)
````

Uma desvantagem do módulo `pickle` é que ele só é capaz de conservar um objeto por vez, e que tem de ser recuperado (desserializado) de uma só vez. Vamos imaginar que o objeto de dados seja um dicionário. Pode ser desejável que não tenhamos de salvar e carregar, a cada vez, o dicionário inteiro, mas salvar e carregar somente um valor único correspondente a uma dada chave. O módulo `shelve` (prateleira) é a solução para esta situação. 'Uma prateleira' - como usado no módulo `shelve` - é um objeto persistente, parecido com um dicionário. A diferença com bancos de dados **dbm** é que os valores (não as chaves!) em uma prateleira podem ser objetos Python arbitrários - algo que o módulo `pickle` possa tratar. Isto inclui a maior parte de exemplos de classe, tipos de dados recursivos e objetos que contêm sub-objetos compartilhados. As chaves têm de ser do tipo ***string***.

O módulo `shelve` pode ser facilmente usado, tal como um dicionário Python. Antes de usarmos um objeto de **prateleira**, precisamos importar o módulo. Depois, abrimos um objeto `shelve` com o método `open()`:

````python
import shelve 
s = shelve.open('MinhaPrateleira')
````

Se o arquivo 'MinhaPrateleira' já existe, o método `open()` tentará abri-lo. Se ele não for um arquivo de prateleira, - isto é, um arquivo criado com o módulo `shelve` - então receberemos uma mensagem de erro. Se o arquivo não existir, ele será criado.

Podemos usar o objeto `s` como um dicionário ordinário, se usarmos chaves como *strings*, e lembrando, que todo objeto de prateleira deve ser fechado com o método `close()`:

````python
s["rua"] = "Av. Goiás"
s["cidade"] = "Goiânia"
for chave in s:
    print(chave)
````

Podemos usar o arquivo de prateleira criado anteriormente, em outro programa ou em uma outra sessão Python interativa:

````python
dic = shelve.open("MinhaPrateleira")
print(dic)
print(dic["rua"], dic["cidade"])
````

Também é possível converter um **objeto de prateleira** num dicionário ordinário com a função `dict()`:

````python
dict(dic)
````

O exemplo seguinte usa valores mais complexos como objeto de prateleira:

````python
fone = shelve.open("AgendaTelefones")
fone["José"] = {"nome":"José", "sobrenome":"Miller", "fone":"99981-4689"}
fone["Cido"] = {"nome":"Aparecido", "sobrenome":"Silva", "fone":"99945-8745"}
fone["Eva"] = {"nome":"Eva", "sobrenome":"Machado", "fone":"99211-9069"}
print(fone["Eva"]["fone"])
fone.close()
````

````python
def bib_padrao(versao=None, so_nivel_superior=True):
    import re   # módulo da bibl. padrão p/ análise de exprs. regulares
    from urllib.request import urlopen
    if versao is None:
        import sys # outro módulo da bibl. padrão
        versao = sys.version_info
        versao = f"{versao.major}.{versao.minor}"
    url = f"https://docs.python.org/{versao}/py-modindex.html"
    with urlopen(url) as f:
        paginaWeb = f.read()
    modulos = set()
    n = 0
    for modulo in re.findall(r'#module-(.*?)[\'"]',
                             paginaWeb.decode('ascii', 'replace')):
        if so_nivel_superior:
            modulo = modulo.split(".")[0]
        modulos.add(modulo)
        n += 1
    return (modulos,n,versao)

mod, qtd, vs = bib_padrao()
print(f"Qtde de módulos da biblioteca padrão da Ling. Python Vs. {vs}: {qtd}\nMódulos:\n{sorted(mod)}")
````

## <a id="anexo2"> Anexo 2 - Erro de Codificação de Arquivos em Python</a>

Um erro comum que ocorre na manipulação de arquivos sequenciais é o da codificação do conteúdo do arquivo, ou seja, o conteúdo do arquivo pode ter sido codificado com um código diferente do usado correntemente...

A menssagem emitida é mostrada a seguir:
``` python
Traceback (most recent call last):  
   File "SCRIPT LOCATION", line NUMBER, in <module>  
     text = file.read()` 
   File "C:\Python31\lib\encodings\cp1252.py", line 23, in decode  
     return codecs.charmap_decode(input,self.errors,decoding_table)[0]
UnicodeDecodeError: 'charmap' codec can't decode byte 0x90 in position 2907500: character maps to `<undefined>`  
```

A páginda de código padrão usado no Windows é a CP1252: ASCII/USA. Se o arquivo tiver sido codificado com outra página, ou se a página ativa no sistema (terminal) for outra, então pode haver conflito de códigos ao tratar o conteúdo de um arquivo... 

Os códigos mais comuns no Brasil são __Latin-1__ e __UTF-8__. 
A correção para o erro indicado passa pela identificação do código usado na geração do arquivo e o ajuste da página de código do sistema para o código usado no arquivo.

Basta especificar a codificação ao abrir o arquivo:
``` python 
arq = open("nome_arq", encoding="código")   # "utf8", "latin-1", ... 
```

````python
nome_arq = "dat/qq.txt" 
with open(nome_arq, encoding="utf-8") as arq:
    texto = arq.read()
print(texto)
````

````python
with open(nome_arq, encoding="latin-1") as arq:
    texto = arq.read()
print(texto)
````

## Tipos de Arquivos

* __Arquivo Texto__ (liso, _plain_): contém texto simples, sem formatação especial. São usados com diversos fins, como armazenar uma versão de um documento, uma cópia de um e-mail em HTML ou um recibo de uma empresa.
* __Arquivo Flat__ (.csv _Comma-separated values_, .tsv - _Tab-separated values_):  arquivo simples/plano, geralmente no formato de texto plano (_plain text_) com valores separados por vírgula/tabulação, usados para armazenar dados de uma estrutura tabular, por exemplo, tabela de banco de dados (BD) ou dados de planilha em uma forma que permite a troca de informações entre BD's. Cada registro na tabela é representado por uma linha do arquivo de texto.
* __Arquivo Matlab__ (.mat): arquivo binário do MATLAB® que armazena variáveis da área de trabalho.
* __Arquivo de Planilha__ (.xls, .xlsx): Extensões de arquivo criado pelo aplicativo **Microsoft Excel** (_Microsoft Excel Open XML Spreadsheet - XLSX_) ou similar. Pode-se abrir esse formato em outros aplicativos de planilha, como **Apple Numbers**, **Google Docs** e **LibreOffice**. São armazenados de forma compactada (arquivos Zip).
* __Arquivo JSON__ (.json): basicamente trata-se de uma longa lista em Python, em que cada item é um dicionário, por exemplo, com quatro chaves - nome do país, código do país, ano e quantidade de habitantes.

### Manipulando Datas
Vamos acrescentar datas ao nosso gráfico para torná-lo mais útil. Os dados referentes às datas são lidos como strings do tipo `yyyy-mm-dd`. Precisamos converter a string `"2018-07-01"` em um objeto que represente esta data. Podemos criar um objeto representando `01 de Julho de 2018` usando o método `strptime()` do módulo `datetime`.

```python
from datetime import datetime
data1 = datetime.strptime('2018-07-01', '%Y-%m-%d')  # segundo argumento indica a formatação desejada para a data
print(data1)
```

Formatos para datas e horários admitidos pelo método `strptime()`:

Argumento | Significado
--------|------------
%A | nome do dia da semana, tal como Segunda-feira
%B | nome do mês, tal como Janeiro
%m | mês, como um número (01 a 12)
%d | dia do mês, como um número (01 a 31)
%Y | ano com quatro dígitos, tal como 2015
%y | ano com dois dígitos, tal como 15
%H | Hora, formato 24h (00 a 23)
%I | Hora, formato 12h (01 a 12)
%p | am ou pm
%M | Minutos (00 a 59)
%S | Segundos (00 a 59)

```python
# pegando datas, temperaturas diárias máximas e mínimas
import csv
nomearq = 'dat\\death_valley_weather_2018_simple.csv'
with open(nomearq) as f:
    leitor = csv.reader(f)
    linha_cab = next(leitor)                      # retorna a próxima linha no arquivo: a primeira linha (contém o cabeçalho)
    datas, maximas, minimas = [], [], []
    print("Temperaturas máximas (°F):")
    for linha in leitor:
        try:
            data_corrente = datetime.strptime(linha[2],'%Y-%m-%d')
            maximas.append(int(linha[5]))             # no arquivo a informação é uma string --> int()
            minimas.append(int(linha[6]))             # no arquivo a informação é uma string --> int()
        except ValueError:
            print(data_corrente, "dados ausentes!")
        else:
            datas.append(data_corrente)
    # print(maximas)
    
# Plotando dados
fig = plt.figure(dpi=128, figsize=(10, 4))
plt.plot(datas, maximas, c='red')
plt.plot(datas, minimas, c='blue')
plt.fill_between(datas, maximas, minimas, facecolor='blue', alpha=0.1)
# Formata o gráfico
plt.title("Temperaturas Máximas Diárias, Jul/2018\nDeath Valley, CA", fontsize=16)
plt.xlabel('datas', fontsize=12); plt.ylabel("Temperatura (°F)", fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)
fig.autofmt_xdate()
```

### Códigos de País com dois caracteres
Antes de nos concentrarmos no mapeamento, precisamos abordar um último aspecto dos dados. A ferramenta de mapeamento em no módulo **Pygal** espera dados em formato específico: os países precisam ser fornecidos como códigos (caracteres) e as populações como valores (numéricos).
Vários conjuntos padronizados de códigos de país são frequentemente usados ao trabalhar com dados geopolíticos; os códigos incluídos no arquivo `population_data.json` são códigos de três letras. Precisamos de uma maneira para encontrar o código de dois caracteres do nome do país.  

Os códigos de país do **Pygal** são armazenados no submódulo `i18n` (abreviação de
*internacionalization* - primeira e última letras e o número de letras entre elas). O dicionário `COUNTRIES` desse módulo contém o código do país com duas letras como chaves e os nomes dos países como valores.

```python
from pygal_maps_world.i18n import COUNTRIES
for cod_pais in sorted(COUNTRIES.keys()):
    print(cod_pais, COUNTRIES[cod_pais])
```

Para extrair os dados do código do país, vamos escrever uma função que pesquisa no dicionário COUNTRIES e retorna o código do país. Vamos escrever isso em um módulo separado chamado `country_codes` para que possamos importá-lo posteriormente para um programa de visualização:

```python
from pygal_maps_world.i18n import COUNTRIES

def get_country_code(nome_pais):
    """Retorna o código de duas letras Pygal para o país 'nome_pais'."""
    for codigo, nome in COUNTRIES.items():
        if nome == nome_pais:
            return codigo
    # se o país não for encontrado, retorna None
    return None

print(get_country_code('Andorra'))
print(get_country_code('United Arab Emirates'))
print(get_country_code('Brazil'))
print(get_country_code('Biscoito Duro'))
```

### Construindo um Mapa do Mundo
Com os códigos de país que temos, é rápido e simples fazer um mapa do mundo. `Pygal` inclui um tipo de gráfico *Worldmap* para ajudar a mapear conjuntos de dados globais. Vamos ver um exemplo de como usar o *Worldmap*, no qual criaremos um mapa simples que destaca a América do Norte, América Central e América do Sul:

```python
import pygal

wm = pygal.maps.world.World() #pygal.Worldmap()
wm.title = 'Américas - Norte, Central e Sul'

wm.add('América do Norte', ['ca', 'mx', 'us'])
wm.add('América Central', ['bz', 'cr', 'gt', 'hn', 'ni', 'pa', 'sv'])
wm.add('América Sul', ['ar', 'bo', 'br', 'cl', 'co', 'ec', 'gf',
'gy', 'pe', 'py', 'sr', 'uy', 've'])
wm.render_to_file('americas.svg')
```
